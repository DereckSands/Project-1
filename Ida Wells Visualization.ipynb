```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('Data Science - Ida_wells.csv', header = None)
```

```python
df.head()
```

```python
df.columns=['Crime','Date','Name','City','State','null1','null2']
```

```python
df.head()
```

```python
del df["null1"]
del df['null2']
```

```python
df.head()
```

```python
df = df.iloc[1:]
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
pltX = df.loc[:, "State"]
pltY = df.loc[:,"Crime"]
figure(figsize=(20,20), dpi=80)
plt.scatter(pltX,pltY, color = "blue", Label = "CRIMES")
```

```python
dic = {}
for column in df['State']:
    if column not in dic:
        dic[column] = 0
    else:
        dic[column]+=1
    print(column)
```

```python
print(dic)
```

```python
states = list(dic.keys())
num = list(dic.values())
figure(figsize=(20,20), dpi=80)


plt.bar(range(len(dic)), num, tick_label=states)
plt.show()
```
